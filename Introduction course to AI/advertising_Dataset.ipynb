```python
import pandas as pd
publi = pd.read_csv('Advertising.csv',index_col=0)
publi.head()
```

```python
print(type(publi))
print(publi.shape)
```

```python
x = publi[['TV','Radio','Newspaper']]
y = publi['Sales']
print(x.shape)
print(y.shape)
```

```python
#Implementação da regressão linear
import seaborn as sns
sns.pairplot(publi,x_vars=['TV','Radio','Newspaper'],y_vars='Sales', size=7,kind='reg')
```

```python
from sklearn.model_selection import train_test_split
# separação dos dados para treinamento e teste
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.3)
```

```python
# Importação do modelo de regressão linear
from sklearn.linear_model import LinearRegression
reglin = LinearRegression()
reglin.fit(xtrain, ytrain)
```

```python
list(zip(['TV','Radio','Newspaper'],reglin.coef_))
```

```python
y_prev = reglin.predict(xtest)
print(y_prev)
print(ytest)
```
